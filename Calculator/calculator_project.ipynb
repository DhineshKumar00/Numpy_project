```python
import numpy as np
```

```python
def addition(a,b):
  try:
    return np.add(a,b)
  except Exception as e:
    print(f"{e} : both array need to be in same shape")

def sub(a,b):
  try:
    return np.subtract(a,b)
  except Exception as e:
    print(f"{e} : both array need to be in same shape")

def multiply(a,b):
  try:
    return np.multiply(a,b)
  except Exception as e:
    print(f"{e} : both array need to be in same shape")

def division(a,b):
  try:
    return np.divide(a,b)
  except ZeroDivisionError:
    print(f"{e} Divide by zero is not possible")
  except Exception as e:
    print(f"{e} : both array need to be in same shape")

def power_array(a, b):
    try:
        return np.power(a, b)
    except ValueError:
        print(" Error: Arrays must be of the same shape for power operation.")
    except OverflowError:
        print(" Warning: Result too large to represent.")


def array_input(name):
   try:
        values = input(f"Enter numbers for {name} separated by spaces: ")
        arr = np.array(list(map(float, values.split())))
        return arr
   except ValueError:
    print("Enter only numeric value")

def Arithmetic_operation():

  print("Enter any operation that you need to perform Arithmetic Operation")
  print("1-Add 2-Sub  3-Multiply  4-Divide 5-Power")

  try:
    array_1=array_input("array 1")
    array_2=array_input("array 2")

    if array_1 is None or array_2 is None:
            print("Exiting due to invalid input.")
            return

    choice = int(input("Enter the operation to do (1-5) - - - "))
    option = {
        1:("Addition",addition),
        2:("Subract",sub),
        3:("Multiply",multiply),
        4:("Divide",division),
        5:("Power",power_array)
    }
    if choice in option:
      name,fun=option[choice]
      print(f"\n Performing {name} \n")
      res = fun(array_1,array_2)

      if res is not None:
        print("Result: ",res)

    else:
      print("Enter Valid option")
  except Exception as e:
    print(f"unExpected Error {e}")

def find_mean(a):
  try:
    return np.mean(a)
  except Exception as e:
    print(f"Unexpected Error : {e}")

def find_median(a):
  try:
    return np.median(a)
  except Exception as e:
    print(f"Unexpected Error : {e}")

def find_sd(a):
  try:
    return np.std(a)
  except Exception as e:
    print(f"Unexpected Error : {e}")

def find_varience(a):
  try:
    return np.var(a)
  except Exception as e:
    print(f"Unexpected Error : {e}")

def find_max(a):
  try:
    return np.max(a)
  except Exception as e:
    print(f"Unexpected Error : {e}")

def find_min(a):
  try:
    return np.min(a)
  except Exception as e:
    print(f"Unexpected Error : {e}")

def find_sum(a):
  try:
    return np.sum(a)
  except Exception as e:
    print(f"Unexpected Error : {e}")

def find_product(a):
  try:
    return np.prod(a)
  except Exception as e:
    print(f"Unexpected Error : {e}")

def find_shape(a):
  try:
    return a.shape
  except Exception as e:
    print(f"Unexpected Error : {e}")

def find_size(a):
  try:
    return a.size
  except Exception as e:
    print(f"Unexpected Error : {e}")

def find_datatype(a):
  try:
    return a.dtype
  except Exception as e:
    print(f"Unexpected Error : {e}")


def Statistical_operation():
  print("Enter the option need to perform in Statistical_operation ")
  print("1-Mean 2-Median 3-Standart Deviation 4-Varience 5-Maximum 6-Minimum \n 7-Sum of arr 8-Product of array 9-Shape 10-size 11-Data type")

  try:
    array_1=array_input("array 1")

    if array_1 is None:
      print("Exiting due to invalid input.")
      return

    choice = int(input("Enter the operation to do (1-11) - - - "))
    option = {
        1:("Mean",find_mean),
        2:("Median",find_median),
        3:("Standard Deviation",find_sd),
        4:("Varience",find_varience),
        5:("Maximum Element",find_max),
        6:("Minimum Element",find_min),
        7:("sum of array",find_sum),
        8:("Product of Array",find_product),
        9:("Shape of Array",find_shape),
        10:("Size of Array",find_size),
        11:("Data Type",find_datatype)
    }
    if choice in option:
      name,fun=option[choice]
      print(f"\n Performing {name} \n")
      res = fun(array_1)

      if res is not None:
        print("Result: ",res)

    else:
      print("Enter Valid option")
  except Exception as e:
    print(f"unExpected Error {e}")


# Manipulation operations

def reshape_arr(a):
  #print("To Reshape array size of one dimensional array must be greater than 4")
  shape_input=input("Enter shape to reshape an array")
  shape = tuple(map(int,shape_input.split(',')))
  try:
    if np.prod(shape) != a.size:
      raise ValueError("The total size of the new shape must match the array size.")

    return a.reshape(shape)

  except ValueError as e:
    print(f"{e} Error")

def Sort_arr(a,order):
  try:
    if order == "asc":
      np.sort(a)
    else:
      np.flip(np.sort(a))
  except Exception as e:
    print(f" Error {e}")

def Unique_value(a):
  try:
    return np.unique(a)
  except Exception as e:
    print(f"Error --- {e}")

def First_Last_index(a):
  try:
    return f"First Index {np.flat[0]}, Last Index {np.flat[-1]}"

  except Exception as e:
    print(f"Error : {e}")

def greater_mean(a):
  try:
    mean=np.mean(a)
    greater_than_mean = a[a>mean]
    return greater_than_mean

  except Exception as e :
    print(f"Error {e}")

def Manipulating_array():
  print("Enter the option need to perform in Manipulation Operation ")
  print("1-ReShape 2-Sorting 3-Unique Values 4-VariFirst and Last index Valuesence 5-Greater Mean ")

  try:
    array_1=array_input("array 1")

    if array_1 is None:
      print("Exiting due to invalid input.")
      return

    choice = int(input("Enter the operation to do (1-5) - - - "))
    option = {
        1:("ReShape",reshape_arr),
        2:("Sorting",Sort_arr),
        3:("Unique Values",Unique_value),
        4:("First and Last index Values ",First_Last_index),
        5:("Greater Mean",greater_mean)

    }
    if choice in option:
      name,fun=option[choice]
      print(f"\n Performing {name} \n")
      res = fun(array_1)

      if res is not None:
        print("Result: ",res)

    else:
      print("Enter Valid option")
  except Exception as e:
    print(f"unExpected Error {e}")

import sys

def main():
  while True:
    print("====================================================")
    print("                 Numpy Calculator                   ")
    print("====================================================")
    print("1-Arithmetic Operation")
    print("2-Statistical Operation")
    print("3-Manipulating Operation")
    print("4-Exit1")



    try:
        choose = int(input("Enter Values btw 1-4 to do operations"))
        op = {
            1:("Arithmetic Operation", Arithmetic_operation),
            2:("Statistical Operation", Statistical_operation),
            3:("Manipulation Operation", Manipulating_array),
            4:("Exit", None)


        }

        if choose in op:
          if choose == 4:
            print("Exiting program. ")
            break

          name,fun=op[choose]
          print(f"\n Performing *** {name} *** \n")
          res = fun()

          if res is not None:
            print("Result: \n",res)

        else:
          print("Enter Valid option")
    except Exception as e:
      print(f"unExpected Error {e}")

    cont = input("\nPress Enter to continue or type 'exit' to quit: ")
    if cont.lower() == 'exit':
      print("Exiting program.")
      break


main()

```
